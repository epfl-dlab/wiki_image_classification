```python
import pickle

import networkx as nx
import numpy as np
import pandas as pd
from tqdm import tqdm

tqdm.pandas()

import sys

sys.path.append("../../../")
from src.config import *
from src.taxonomy.heuristics import Heuristics
```

```python
n_chunks = 100
embeddings_dict = {}
for chunk in tqdm(range(n_chunks)):
    # for chunk in tqdm(range(1)):
    with open(EMBEDDINGS_PATH + f"chunk{chunk}.pkl", "rb") as fp:
        embeddings_dict.update(pickle.load(fp))
```
